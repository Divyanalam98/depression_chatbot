```python
import openai
openai.api_key = <YOUR-KEY>
```

```python
model_engine = "davinci:ft-san-jose-state-university-2023-10-04-05-08-41"

# Define prompt and reference text
prompt = "Are there different types of depression? If so, what are they? ->	"
reference = "Different types of depression include major depressive disorder, persistent depressive disorder, seasonal affective disorder, postpartum depression, bipolar disorder, and psychotic depression."

# Generate text using OpenAI model
response = openai.Completion.create(
    engine=model_engine,
    prompt=prompt,
    temperature=0.2,
    max_tokens=100,
    n=1
)
generated = response.choices[0].text.strip() +"."

# Calculate cosine similarity between generated text and reference text
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([generated, reference])
similarity = cosine_similarity(vectors)[0][1]

# Print results
print("Generated text:", generated)
print("Reference text:", reference)
print("Cosine similarity:", similarity)

```

```python
!pip install rouge-score
from rouge_score import rouge_scorer

# Calculate Rouge scores between generated text and reference text
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(reference, generated)

# Print results
print("Generated text:", generated)
print("Reference text:", reference)
print("Rouge-1 score:", scores['rouge1'].fmeasure)
print("Rouge-2 score:", scores['rouge2'].fmeasure)
print("Rouge-L score:", scores['rougeL'].fmeasure)
```

```python
import sacrebleu
# Calculate BLEU score between generated text and reference text
score = sacrebleu.corpus_bleu([generated], [[reference]]).score

# Print results
print("Generated text:", generated)
print("Reference text:", reference)
print("BLEU score:", int(score))
```

```python
model_engine = "davinci"

# Define prompt and reference text
prompt = "Are there different types of depression? If so, what are they? ->	"
reference = "Different types of depression include major depressive disorder, persistent depressive disorder, seasonal affective disorder, postpartum depression, bipolar disorder, and psychotic depression."

# Generate text using OpenAI model
generated = openai.Completion.create(
    engine=model_engine,
    prompt=prompt,
    temperature=0.2,
    max_tokens=100,
    n=1
)["choices"][0]["text"]

# Calculate cosine similarity between generated text and reference text
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([generated, reference])
similarity = cosine_similarity(vectors)[0][1]

# Print results
print("Generated text:", generated)
print("Reference text:", reference)
print("Cosine similarity:", similarity)

```

```python

# Calculate Rouge scores between generated text and reference text
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(reference, generated)

# Print results
print("Generated text:", generated)
print("Reference text:", reference)
print("Rouge-1 score:", scores['rouge1'].fmeasure)
print("Rouge-2 score:", scores['rouge2'].fmeasure)
print("Rouge-L score:", scores['rougeL'].fmeasure)
```

```python
import sacrebleu

# Calculate BLEU score between generated text and reference text
score = sacrebleu.corpus_bleu([generated], [[reference]]).score


# Print results
print("Generated text:", generated)
print("Reference text:", reference)
print("BLEU score:", int(score))
```
